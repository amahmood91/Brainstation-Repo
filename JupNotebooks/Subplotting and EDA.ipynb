```python
# For Plotting
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Manipulating and Storing Data
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)

# Machine Learning
from sklearn import linear_model
```

```python
df = pd.read_csv('data/melb_data.csv')
df.head(3)
```

```python
print("Unique Values by Column")
for col in df.columns:
    print(f"{col}: {len(df[col].unique())}")
```

```python
categorical_columns = [col for col in df.columns if len(df[col].unique())<20]
categorical_columns
```

```python
# Label encoding RegionName to avoid overly verbose outputs...we will cover this later.
df['Regionname'] , _ = pd.factorize(df['Regionname'])
```

```python
f,axes = plt.subplots(2,4,figsize=(30,20))
axes = axes.ravel()
for i,col in enumerate(categorical_columns):
    sns.countplot(df[col],ax=axes[i])
    axes[i].set_title(f'{col}',size=20)
    axes[i].set_xlabel(col,size=15)
    axes[i].set_ylabel('Counts',size=15)
    axes[i].tick_params(axis='y', labelsize=14)
plt.suptitle('Distributions of Categorical Variables',
             size=30,
             y=0.95)
                            
plt.show()
```

Rooms and Bedroom2 appear to be duplicates of each other so we can drop one of these columns. From the distributions we can see a few things right away. Most houses have three bedrooms and 1-2 bathrooms. A vast majority of houses contain 1-2 cars (likely parking spaces).

```python
df = df.drop(columns=['Bedroom2'])
```

# Numeric Plots and Suspicious Values

```python
numeric_columns = ['Price','Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude']
```

```python
f, axes = plt.subplots(3,2, figsize=(30,20))
axes = axes.ravel() #takes the 3 x 2 matrix and turns it into a 6 element object (reduces dimensionality)

for i, col in enumerate(numeric_columns):
    sns.distplot(df[col].dropna(), ax=axes[i])
    axes[i].set_title(col)
plt.show()


```

```python
#we see some of our plots need to have a limited domain

case = df['Landsize'] < df['Landsize'].median()*4  #this will give a boolean array

sns.distplot(df[case]['Landsize'].dropna())
```

```python
#we want to get rid of this 0 value
df['Landsize'] = df['Landsize'].replace(0, np.nan)  #replaces 0 with NANs
case = df['Landsize'] < df['Landsize'].median()*4  #this will give a boolean array

sns.distplot(df[case]['Landsize'].dropna())
```

```python
#Lets look at BuildingArea too
case = df['BuildingArea'] < df['BuildingArea'].median()*4  #this will give a boolean array
sns.distplot(df[case]['BuildingArea'].dropna())
```

# Multi and Bivariate Analysis

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
corr = df.corr()['Price']
sns.barplot(x=corr.values, y = corr.index)
corr
```

```python
corr = corr.sort_values(ascending=False)[1:] #we know price will correlate with itself, so remove it from this plot - i.e plot at index 1 and on (ignores Price)
sns.barplot(corr.values, corr.index)
```

```python
#to assess predicted power
corr = np.abs(corr).sort_values(ascending=False)
sns.barplot(corr.values, corr.index)
```
