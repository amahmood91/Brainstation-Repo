## Load in the Data

````python
data_df = pd.read_csv('projectdata/data.csv')
data_df.head()
````

We want to create a bar plot showing how many counties are in each state.

````python
# Create a series with the count
county_count = data_df.State.value_counts()
county_count.head()
````

````python
# These are the values we want for the x-values
county_count.index
````

````python
# These are the values we want for the y-values
county_count.values
````

## Creating Bar Plot

````python
# bar plot

#Create a figure size 10in wide and 5in tall.
plt.figure(figsize=(10,5))

plt.bar(county_count.index, county_count.values)

#Rotate the x-ticks 90 degrees so it doesn't overlap with each other
plt.xticks(rotation = 90, fontsize = 8)

# Put in titles & axes labels
plt.title('Number of Counties by State')
plt.xlabel('State')
plt.ylabel('Counties')

# Make sure everything spaced out better in the figure
plt.tight_layout()

# Save the figure as png
plt.savefig('State_count_bar.png')
````
