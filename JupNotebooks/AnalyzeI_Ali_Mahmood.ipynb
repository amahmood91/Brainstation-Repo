#### Create a model which measures the fraction of votes a party (Republican) gets within a county based on the county data. When building your model keep the following points in mind:

#### 1. What are the data points? What are the dependent and independent variables?
#### 2. We are asking you to fit the fraction of votes that the Republican party gets within a county, this value is any continuous value between 0 and 1. What model is appropriate for this sort of problem?

The data points will be the county-level data in the main_df dataframe. The dependent variable is the Republican Voter Share and the independent variables, or predictors, are the following: "White (Not Latino) Population" and "African American Population". Although many variables may play a factor, these two population subsets have been shown in previous figures to be the most strongly correlated with the outcome (Republican Voter share).


Since our prediction(s) will be a continuous output, a linear regression model - particualrly multiple linear regression would be appropriate since we have a number of different predictor variables.

```python
from statsmodels.regression import linear_model
import statsmodels.api as sm
```

```python
#function to calculate R2
def calc_R2(model, y):
    ss_resid = np.sum((model.fittedvalues - y)**2)
    ss_total = np.sum((y - np.mean(y))**2)
    return round(1 - ss_resid/ss_total, 4)
```

```python
#Set dependent var () and Independent vars (X)

#independent variables
ind_vars = ['White (Not Latino) Population', 'African American Population']


#No regularizatopn
X = sm.add_constant(main_df[ind_vars].values)
#y = main_df['Rep_08_Ratio'].values
#y = main_df['Rep_12_Ratio'].values
y = main_df['Rep_16_Ratio'].values
model_votes = sm.OLS(y,X).fit()
```

```python
#View summary
print(model_votes.summary())

#Predict
beta0_x = 1
x_1 = 68
x_2 = 28
predicted_Rep_ratio = round(model_votes.predict([beta0_x,x_1, x_2])[0],2)

print(f"\nGiven a 'White (Not Latino) Population' of {x_1} and an African American Population of {x_2},\nthe model predicts a Republican Vote Share of {predicted_Rep_ratio}% with an R squared value of {calc_R2(model_votes,y)}.")
```

```python
#Assess model to see which method of regularization to apply (if it provides a higher R squared value than the unfitted model)
plain_model = sm.OLS(y,X).fit()
print('Plain model results:')
print(f"Beta 0: {plain_model.params[0]}, Beta 1: {plain_model.params[1]}, Beta 2: {plain_model.params[2]}")
print(f'R2: {calc_R2(plain_model, y)}')

L1_model = sm.OLS(y,X).fit_regularized(alpha = 1, L1_wt = 1)
print('\nL1 model results:')
print(f"Beta 0: {L1_model.params[0]}, Beta 1: {L1_model.params[1]}, Beta 2: {L1_model.params[2]}")
print(f'R2: {calc_R2(L1_model, y)}')

L2_model = sm.OLS(y,X).fit_regularized(alpha = 1, L1_wt = 0)
print('\nL2 model results:')
print(f"Beta 0: {L2_model.params[0]}, Beta 1: {L2_model.params[1]}, Beta 2: {L2_model.params[2]}")
print(f'R2: {calc_R2(L2_model, y)}')
```

We see that regularization lowers the R squared which could mean soem overfitting was corrected. Assess the accuracy of the their predictions

```python
#First, create a copy of the data but with just the independent and dependant variable columns

assess_df = main_df[ind_vars + ['Rep_08_Ratio', 'Rep_12_Ratio', 'Rep_16_Ratio']]
assess_df.head()
```

```python
#Put predictions into columns
assess_df['Rep_Ratio_PRED_plain'] = assess_df.apply(lambda row_val: 
                                      round(plain_model.predict([1,row_val['White (Not Latino) Population'], row_val['African American Population']])[0],2),
                                      axis=1)

assess_df['Rep_Ratio_PRED_L1'] = assess_df.apply(lambda row_val: 
                                      round(L1_model.predict([1,row_val['White (Not Latino) Population'], row_val['African American Population']])[0],2),
                                      axis=1)

assess_df['Rep_Ratio_PRED_L2'] = assess_df.apply(lambda row_val: 
                                      round(L2_model.predict([1,row_val['White (Not Latino) Population'], row_val['African American Population']])[0],2),
                                      axis=1)

```

```python
#2016 actual vs plain model predictions
plt.figure(figsize=(15,5))

plt.hist(assess_df['Rep_Ratio_PRED_L2'].values, bins = 100, label = 'L2 Prediction', alpha = 0.5)
plt.title('Republican Voter Share across Counties')
plt.xlabel('Republican Voter Share')
plt.ylabel('Number of Counties')

plt.hist(assess_df['Rep_Ratio_PRED_L1'].values, bins = 100, label = 'L1 Prediction', alpha = 0.5)
plt.title('Republican Voter Share across Counties')
plt.xlabel('Republican Voter Share')
plt.ylabel('Number of Counties')

plt.hist(assess_df['Rep_Ratio_PRED_plain'].values, bins = 100, label = 'Plain Prediction', alpha = 0.5)
plt.title('Republican Voter Share across Counties')
plt.xlabel('Republican Voter Share')
plt.ylabel('Number of Counties')

plt.hist(assess_df['Rep_16_Ratio'].values, bins = 100, label = '2016 Actual', alpha = 0.5)
plt.title('Republican Voter Share across Counties')
plt.xlabel('Republican Voter Share')
plt.ylabel('Number of Counties')


plt.legend()

plt.show()
```

Since our independent variables are static (i.e there is no 2008 population data, 2012 population data - we just have population data...) we can take the average vote share ratio between them which will allow us to fit the model on both years to eventualy predict the 2016 election

```python
#Get the mean of the vote share ratios of 2008 and 2012 elections
model_df['Rep_Ratio_08-12_avg'] = model_df[['Rep_08_Ratio','Rep_12_Ratio']].mean(axis=1)
model_df.head()

#now lets predict the 2016 Rep ratio using the Rep_Ratio_08_12_avg column as our dependent var and White (Not Latino) Population and African American Population as our two independent vars


#No regularizatopn
X = sm.add_constant(model_df[ind_vars].values)
y = model_df['Rep_Ratio_08-12_avg'].values
model_votes_2 = sm.OLS(y,X).fit()

#Put 2016 predictions into columns
model_df['Rep_Ratio_16_PRED'] = model_df.apply(lambda row_val: 
                                      round(model_votes_2.predict([1,row_val['White (Not Latino) Population'], row_val['African American Population']])[0],2),
                                      axis=1)

model_df.head()
```

Now we have a column of predicted Republican vote share for the 2016 election per county BASED on the outcomes of the 2008 and 2012 data

```python
#Get Raw votes of predicted 2016 vote share
model_df['Rep_Raw_votes_08-12_PRED'] = round(model_df.apply(lambda row_val: 
                                      row_val['Total Population']*(row_val['Rep_Ratio_16_PRED']/100),
                                      axis=1))
model_df['Dem_Raw_votes_08-12_PRED'] = round(model_df.apply(lambda row_val: 
                                      row_val['Total Population'] - row_val['Rep_Raw_votes_08-12_PRED'],
                                      axis=1))

#FOR COMPARISON - Get Raw votes of actual 2016 vote share
model_df['Rep_Raw_votes_08-12_ACT'] = round(model_df.apply(lambda row_val: 
                                      row_val['Total Population']*(row_val['Rep_16_Ratio']/100),
                                      axis=1))
model_df['Dem_Raw_votes_08-12_ACT'] = round(model_df.apply(lambda row_val: 
                                      row_val['Total Population'] - row_val['Rep_Raw_votes_08-12_ACT'],
                                      axis=1))


model_df.head()
```

```python
#Lets seperate the actual data from the predicted to make things more organized
model_pred_df = model_df[['State','Rep_Raw_votes_08-12_PRED','Dem_Raw_votes_08-12_PRED']]
display(model_pred_df.head())

model_actual_df = model_df[['State','Rep_Raw_votes_08-12_ACT','Dem_Raw_votes_08-12_ACT']]
display(model_actual_df.head())
```

```python
#Now lets get the total votes per State
#predictions
state_pred_df = model_pred_df.groupby('State').sum()
#actual
state_act_df = model_actual_df.groupby('State').sum()


#Make a column to indicate who won
state_pred_df['Winner'] = state_pred_df.apply(lambda row_val: "Republican" if row_val['Rep_Raw_votes_08-12_PRED'] > row_val['Dem_Raw_votes_08-12_PRED'] else "Democrat", axis = 1)
state_act_df['Winner'] = state_act_df.apply(lambda row_val: "Republican" if row_val['Rep_Raw_votes_08-12_ACT'] > row_val['Dem_Raw_votes_08-12_ACT'] else "Democrat", axis = 1)

#reset the index
state_pred_df.reset_index(inplace=True)
state_act_df.reset_index(inplace=True)


display(state_pred_df.head()) #note that State has become the index
display(state_act_df.head()) #note that State has become the index
```

Now that we can see which party won for each state, lets show how many electoral collage votes they get, and ultimately which party has the most between all states

```python
#Recall that during data cleaning, Alaska and District Of Columbia were removed due to missing data, as such we will do a left join
state_pred_df = pd.merge(state_pred_df, ec_data, on = 'State', how = 'left')
state_act_df = pd.merge(state_act_df, ec_data, on = 'State', how = 'left')

display(state_pred_df.head())
display(state_act_df.head())
```

Now lets see which party is predicted to have the most electoral votes in 2016. We'll also show the actual outcome for comparison

```python
election_outcome_pred = state_pred_df.groupby(['Winner']).sum()['EC_votes']
election_outcome_act = state_act_df.groupby(['Winner']).sum()['EC_votes']

print("Predicted", election_outcome_pred.sort_values(ascending=False))
print("\nActual",election_outcome_act.sort_values(ascending=False))

#Republicans predicted to win election since they have the most Electoral Votes
```

#### The model's prediction that the Republican party wins the 2016 election is correct, although the number of electoral collage votes they won is extremely exaggerated (528 to 4!). This heavy skew is due to the model using only two predictors - White population and the African American population - where Caucasians are the vast majority across all states and African Americans are a minority. Combined with the fact that the model recognized a stronger correlation between Republican votes and White population versus Democrat votes and African American votes, it would stronger favor predicting Republican county and state victories. It should also be notes that voter turn-out by ethnicity is likely a very strong factor in this result as well where turnout is lower among minority groups compared to White populations. A more balanced model prediction could be obtained by correcting for such a heavy dependence on the white population by adding additional independent variables such as Asian American Population, Native American Population, etc... and possibly including other metrics on occupation and education levels.

#### *Things to note about the 'Actual' Winner table above - analysis on the dataset was done under the assumption that there are only two parties "Democrats" and "Republicans". Republican vote share per county (actual) for each election was calculated by dividing the number of Republican votes by the total of Republican and Democrat votes, and subsequently converted to a percentage. Democrat vote share was calculated based on the two-party assumption and as such was the complement of the Republican vote share (i.e if Republicans had 56%, then Democrats had 1 - 0.56 = 44%). This is an important point to take into account when evaulating the outcome of the above analysis as the actual table shows that Democrats won the 2016 election when we know they did not. This discrepency is heavily due to other party votes (such as Independents) splitting the vote in real life - their vote share was not included in the dataset and as such this resulted in infalted "Actual" Democratic vote share in the analysis above.
